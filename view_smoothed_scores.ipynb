### View all of the smoothed college confessions sentiment scores 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
caltech_smoothed = "berkout_smoothed.csv"
df = pd.read_csv(caltech_smoothed, header=0, sep=u",", index_col = None, encoding="ISO-8859-1")
df.head()
```

```python
def plot_metric(score_name, df):
    """
    Helper function to plot a smoothed metric over time 
    """
    timestamps = df["timestamp"].tolist()
    scores = df[score_name].tolist()
    plt.plot(timestamps, scores)
    plt.title("Sentiment " + score_name + " score vs. Timestamp")
    plt.show()
```

```python
plot_metric("compound_smoothed", df)
```

```python
plot_metric("positive_smoothed", df)
```

```python
plot_metric("negative_smoothed", df)
```

```python
plot_metric("neutral_smoothed", df)
```

```python
caltech_smoothed = "caltech_smoothed.csv"
caltech = pd.read_csv(caltech_smoothed, header=0, sep=u",", index_col = None, encoding="ISO-8859-1")
berkeley_smoothed = "berkout_smoothed.csv"
berkeley = pd.read_csv(berkeley_smoothed, header=0, sep=u",", index_col = None, encoding="ISO-8859-1")
mit_smoothed = "mitout_smoothed.csv"
mit = pd.read_csv(mit_smoothed, header=0, sep=u",", index_col = None, encoding="ISO-8859-1")
uci_smoothed = "uciout_smoothed.csv"
uci = pd.read_csv(uci_smoothed, header=0, sep=u",", index_col = None, encoding="ISO-8859-1")
```

```python
caltech_compound = caltech["compound_smoothed"].dropna().tolist()
berkeley_compound = berkeley["compound_smoothed"].dropna().tolist()
mit_compound = mit["compound_smoothed"].dropna().tolist()
uci_compound = uci["compound_smoothed"].dropna().tolist()
print("Caltech confessions size: " + str(len(caltech_compound)))
print("Berkeley confessions size: " + str(len(berkeley_compound)))
print("MIT confessions size: " + str(len(mit_compound)))
print("UCI confessions size: " + str(len(uci_compound)))
```

```python
# Keep track of the indices to mark data points
caltech_range = len(caltech_compound)
berkeley_range = len(berkeley_compound)
mit_range = len(mit_compound)
uci_range = len(uci_compound)

tMax = min(caltech_range, berkeley_range, mit_range, uci_range)
t = np.arange(tMax)
dataset = []
for i in range(tMax):
    
    # Form the datapoint to be put into the html file 
    pointDict = {"x": i}
    pointDict["val_0"] = caltech_compound[i]
    pointDict["val_1"] = berkeley_compound[i]
    pointDict["val_2"] = mit_compound[i]
    pointDict["val_3"] = uci_compound[i]
    
    dataset.append(pointDict)

print(dataset)
```
